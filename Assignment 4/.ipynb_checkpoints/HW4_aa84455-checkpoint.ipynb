## Question 2 - Concepts (10 pts)

Emma is a high school student who is creating a model for spam classification, which is well known to be an imbalanced classification problem. During the course of modeling, she had various observations and she is not sure if these observations suggest anything about the correctness of her modeling exercise. You are given two such instances below. For each observation identify what the problem with it may be (if any) and suggest a way to correct it. 

1. She thinks that the most important signal to identify spam emails is the sender. So, she creates a feature called is_sender_spammer and populates it with the percentage of emails sent by the sender that were spam, in the given dataset. The way she created this feature was - for each email, she counted the number of emails sent by each sender which were labeled as true spam and divided by the total number of emails sent by the sender. She then divides the given dataset into train and test and trains a logistic regression classifier using just this one feature. The AUROC on the test dataset comes out to be 0.99. She thinks this is too good to be true. What do you think may be the problem with her approach, and how can this be fixed?

2. She then changes her approach and tries multiple sets of features with different models. She finds out that whatever features or models she uses the classification accuracy on the test dataset is always above 90%. Why do you think this could happen?


## Part 1:

### The problem

In this case, Emma is creating a predictor variable based on the label that we have for each data point (dependent variable). This way, her creafed feature (% of spam emails) is not truly independent and can be considered as seen for all the users in the test set, making the test data points easily classifiable. She has in a way tampered with the 'unseenness' of her test data since for each user occuring in the test data, she has already seen the % of spam emails during her training set.

This will make her model ungeneralisable since it will fail for previously unseen users when it is deployed (we do not know % of spam emails sent by a new user)

### The solution

Emma should use predictors that can be observed for previously unseen users, predictors that are truly independent of the output. Predictors at an email level will be more suitable than predictors at a user level since they will be observable for even previously unseen users.

eg. Length of email, content of email etc.

## Part 2:

### The problem

Since this is an imbalanced class problem, it is safe to say that either one of the classes is representing >90% of the data. Let us assume that non-spam class is dominant. Due to this imbalance, even if the model classifies every email as non-spam, it will achieve > 90% accuracy without catching even a single spam email.

### The solution

To estimate the ability of her classifier to identify spam emails, Emma should use metrics like f-score (harmonic mean of precision and recall), which will penalise for any of the two being low. Precision and recall capture the goodness of true positive identification, and low values of any of the two will signal poor identification of the spam emails through a low f-score.

```python
#DONOT CHANGE THIS
df_train = df.sample(frac=0.8,random_state=11)
df_test = df.drop(df_train.index)

from sklearn.base import BaseEstimator, TransformerMixin
#We provide the required transformers to build the pipeline.
# Transfomer for performing one hot encoding on a specified categorical column
class OneHotEncoder(BaseEstimator, TransformerMixin):
    
    def __init__(self,col):
        self.col = col
                
    def transform(self, X):
        return pd.get_dummies(X[self.col]) 

    def fit(self, X, y=None):
        return self  
    
# Stateless transformer for selecting a specified columns
class DFSubsetSelector(BaseEstimator,TransformerMixin):
    
    def __init__(self,cols):
        self.cols = cols
        
    def transform(self ,X):
        return X[self.cols]
    
    def fit(self, X, y=None):
        return self
    
class ConcatFeatures(BaseEstimator, TransformerMixin):
    
    def __init__(self,cols,sep="_"):
        self.cols = cols
        self.sep = sep
                
    def transform(self, X):
        concat_col = self.sep.join(self.cols)
        return X.apply(lambda x : self.sep.join([x[col] for col in self.cols]) ,axis=1).to_frame(name = concat_col)

    def fit(self, X, y=None):
        return self  
```

#### (3)

```python
class MinMaxTransform(BaseEstimator, TransformerMixin):
    
    def __init__(self,cols=None):
        self.cols = cols
                
    def transform(self, X):
        return X.apply(lambda x : (x-min(x))/(max(x)-min(x)))

    def fit(self, X, y=None):
        return self  
```

#### (4)

```python
X_train = df_train.drop('Exited', axis = 1)
X_test = df_test.drop('Exited', axis = 1)
Y_train = df_train['Exited']
Y_test = df_test['Exited']
```

```python
numeric_features = X_train.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X_train.select_dtypes(include=['object']).columns
```

```python
#ML Pipeline for Logistic Regression model

from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

numeric_transformer = Pipeline(steps=[
    ('minmax', MinMaxTransform())])
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(categorical_features))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

lr = Pipeline(steps=[('preprocessor', preprocessor),
                     ('classifier', LogisticRegression())])
lr_fit = lr.fit(X_train, Y_train)
predict = lr_fit.predict(X_test)

score = f1_score(Y_test,predict)
auc = roc_auc_score(Y_test,predict)
print('F1 score for Logistic Regression:' , score )
print('AUROC for Logistic Regression:', auc)
```

#### (5)

```python
#Adding 3 new features - Balance_Salary, Tenure_Age, CreditScore_Age
df_train1 = df_train
df_test1 = df_test

df_train1['Balance_Salary'] = df_train1['Balance']/df_train1['EstimatedSalary']
df_train1['Tenure_Age'] = df_train1['Tenure']/df_train1['Age']
df_train1['CreditScore_Age'] = df_train1['CreditScore']/df_train1['Age']

df_test1['Balance_Salary'] = df_test1['Balance']/df_test1['EstimatedSalary']
df_test1['Tenure_Age'] = df_test1['Tenure']/df_test1['Age']
df_test1['CreditScore_Age'] = df_test1['CreditScore']/df_test1['Age']
```

```python
X_train1 = df_train1.drop('Exited', axis = 1)
X_test1 = df_test1.drop('Exited', axis = 1)
Y_train1 = df_train1['Exited']
Y_test1 = df_test1['Exited']

numeric_features = X_train1.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X_train1.select_dtypes(include=['object']).columns

numeric_transformer = Pipeline(steps=[
    ('minmax', MinMaxTransform())])
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(categorical_features))])


preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

lr1 = Pipeline(steps=[('preprocessor', preprocessor),
                     ('classifier', LogisticRegression())])

lr_fit1 = lr1.fit(X_train1, Y_train1)
predict1 = lr_fit1.predict(X_test1)

score = f1_score(Y_test1,predict1)
auc = roc_auc_score(Y_test1,predict1)
print('F1 score for Logistic Regression with new features:' , score )
print('AUROC for Logistic Regression with new features:', auc)
```

```python
lr2 = Pipeline(steps=[('preprocessor', preprocessor),
                     ('classifier', LogisticRegression(class_weight='balanced'))])

lr_fit2 = lr2.fit(X_train1, Y_train1)
predict2 = lr_fit2.predict(X_test1)

score = f1_score(Y_test1,predict2)
auc = roc_auc_score(Y_test1,predict2)
print('F1 score for Logistic Regression with new features and class_weight= balanced option:' , score )
print('AUROC for Logistic Regression with new features and class_weight= balanced option:', auc)
```

#### (6)

```python
import itertools
marker = itertools.cycle((',', '+', '.', 'o', '*', 'v', '1', 's', 'p','d','|','x')) 

# Function to plot ROC curve
def plot_roc_curve(fpr,tpr,label=None,marker= None):
    plt.plot(fpr,tpr,linewidth=2,label=label,marker = marker)
    plt.plot([0,1],[0,1],'k--')
    plt.axis([0,1,0,1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
```

```python
mlp_1 = MLPClassifier(hidden_layer_sizes=(100, ), activation="relu", solver="adam")
mlp_2 = MLPClassifier(hidden_layer_sizes=(50,100,50), activation="relu", solver="adam")
mlp_3 = MLPClassifier(hidden_layer_sizes=(50,50,50), activation="relu", solver="adam")
mlp_4 = MLPClassifier(hidden_layer_sizes=(100, ), activation="tanh", solver="adam")
mlp_5 = MLPClassifier(hidden_layer_sizes=(50,100,50), activation="tanh", solver="adam")
mlp_6 = MLPClassifier(hidden_layer_sizes=(50,50,50), activation="tanh", solver="adam")
mlp_7 = MLPClassifier(hidden_layer_sizes=(100, ), activation="relu", solver="sgd")
mlp_8 = MLPClassifier(hidden_layer_sizes=(50,100,50), activation="relu", solver="lbfgs")
mlp_9 = MLPClassifier(hidden_layer_sizes=(50,50,50), activation="relu", solver="sgd")
mlp_10 = MLPClassifier(hidden_layer_sizes=(100, ), activation="tanh", solver="lbfgs")
mlp_11 = MLPClassifier(hidden_layer_sizes=(50,100,50), activation="tanh", solver="sgd")
mlp_12 = MLPClassifier(hidden_layer_sizes=(50,50,50), activation="tanh", solver="adam")

classifiers = [mlp_1,mlp_2,mlp_3,mlp_4,mlp_5,mlp_6,mlp_7,mlp_8,mlp_9,mlp_10,mlp_11,mlp_12]
classifiers_name = ['mlp1','mlp2','mlp3','mlp4','mlp5','mlp6','mlp7','mlp8','mlp9','mlp10','mlp11','mlp12']
fig, ax = plt.subplots(figsize=(15,10))
i=0
for classifier in classifiers:
    mlp = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', classifier)])
    mlp.fit(X_train1, Y_train1)
    pred = mlp.predict_proba(X_test1)[:,1]
    fpr, tpr, th = roc_curve(Y_test1,pred)    
    plot_roc_curve(fpr,tpr,classifiers_name[i],marker = marker.next())
    i=i+1

plt.legend(loc="lower right")
plt.show()
```

```python
# Best MLPClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV

parameter_space = {
    'classifier__hidden_layer_sizes': [(50,50,50), (50,100,50), (100,)],
    'classifier__activation': ['tanh', 'relu'],
    'classifier__solver': ['sgd', 'adam'],
    'classifier__alpha': [0.0001, 0.05],
    'classifier__learning_rate': ['constant','adaptive'],
}

best_mlp = Pipeline(steps=[('preprocessor', preprocessor),
                           ('classifier', MLPClassifier(max_iter=100))])

lr3 = GridSearchCV(best_mlp, parameter_space, n_jobs= 1)                  
lr3.fit(X_train1,Y_train1)
predict3 = lr3.predict(X_test1)

score = f1_score(Y_test1,predict3)
auc = roc_auc_score(Y_test1,predict3)
print('F1 score for best MLP classifier:' , score )
print('AUROC for best MLP classifier:', auc)
```
